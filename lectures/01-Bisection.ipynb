# Bisection algorithm 

We begin our study of numerical methods with the bisection algorithm for solving the nonlinear equation $f(x) = 0$. 

  * implementation in code
  * convergence rates
  * error estimates
  * stopping conditions
  * computational cost  
  * conditioning (sensitivty)

 

```julia
using Plots
f(x) = x.^2 - 2
a,b = 1,2
c = (a+b)/2

x = linspace(0,2)
plot(x,f(x), xlabel="x", ylabel="f(x)")
plot!([a, b],[0, 0])
```
