# Newton divided differences

In class we worked out that the four datapoints

|| | | | |
|-------|---|---|---|---|
| $x_i$ | 1 | 2 | 3 | 4 |
| $y_i$ | 6 | 9 | 2 | 5 |


are interpolated by the third-order polynomial


\begin{equation*}
P(x) = 6 + (x-1) (3 + (x-2) (-5 + (x-3) 10/3))
\end{equation*}

```julia
xdata = [1, 2, 3, 4]
ydata = [6, 9, 2, 5]
```

```julia
using Plots
gr()
scatter(xdata, ydata, marker=:circ, color="black", xlim=(0,5), ylim=(0,10), label="datapoints")


```

```julia
scatter(xdata, ydata, color="black", label="datapts")
plot!(x, 6 + 0*x, linestyle=:solid, color="blue", label="P0(x)")
```

```julia
P0(x) = 6
P1(x) = 6 + 3*(x-1)
P2(x) = 6 + (x-1)*(3 + (x-2)*(-5))
P3(x) = 6 + (x-1)*(3 + (x-2)*(-5 + (x-3)*10/3))
```

```julia
P2.([1,2])
```

```julia
scatter(xdata, ydata, color="black", label="datapts", xlim=(0,5), ylim=(0,10))
plot!(x, P0.(x), linestyle=:solid, color="blue", label="P0(x)")
plot!(x, P1.(x), linestyle=:solid, color="red", label="P1(x)")
plot!(x, P2.(x), linestyle=:solid, color="green", label="P2(x)")
plot!(x, P3.(x), linestyle=:solid, color="magenta", label="P3(x)")
```

```julia
P2(x) = 6 + (x-1)*(3 + (x-2)*(-5))

scatter(xdata, ydata, color="black", label="datapts", xlim=(0,5), ylim=(0,10))
plot!(x, 6 + 0*x, linestyle=:solid, color="blue", label="P0(x)")
plot!(x, P1(x), linestyle=:solid, color="red", label="P1(x)")
plot!(x, P2(x), linestyle=:solid, color="green", label="P2(x)")
```
